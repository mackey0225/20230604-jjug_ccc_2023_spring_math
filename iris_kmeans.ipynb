```
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
%matplotlib inline
```

```
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = [iris.target_names[i] for i in iris.target]
```

```
df.describe()
```

```
sns.pairplot(df, hue="target")
```

```
df_data = pd.DataFrame(iris.data, columns=iris.feature_names)
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=1, max_iter=10000)
kmeans.fit(df_data)
df_clustered = df_data
df_clustered['cluster'] = pd.Series(kmeans.labels_)
sns.pairplot(df_clustered, hue="cluster")
```
